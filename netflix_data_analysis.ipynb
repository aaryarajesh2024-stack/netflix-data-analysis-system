#2. Data Cleaning

```python
#to handle and fill in the missing values appropriately
#the columns having missing values are director,cast,country,date_added,rating,duration

#to drop all the rows were the values are missing
df.dropna(subset=['date_added'],inplace=True)
df.dropna(subset=['duration'],inplace=True)
df.dropna(subset=['rating'],inplace=True)
df.dropna(subset=['country'],inplace=True)
df.dropna(subset=['director'],inplace=True)
df.dropna(subset=['cast'],inplace=True)

#to check if all null values has been handled properly
df.isnull().sum()

#all null values have been dropped
```

```python
#Standardize casing for string columns like type and rating.
#making the characters of type column in lowercase and that of rating column in uppercase

df["type"]=df["type"].str.lower()
df["rating"]=df["rating"].str.upper()

#to check if the casing has changes

df.head()
```

```python
#to convert the date column into datetime type

df["date_added"]=pd.to_datetime(df["date_added"],format='mixed')

#now to check if the datatype has been converted

df["date_added"].dtype

#the datatype has been changed
```

```python
#creating new columns from date_added column for analysis

df["Year"]=df["date_added"].dt.year
df["Month"]=df["date_added"].dt.month
df["Day"]=df["date_added"].dt.day

#to check if the columns are added
df.head()

#the columns have been added
```

#3. Univariate and Bivariate Analysis

```python
# Identify the most common rating category.

mcommon_rating=df['rating'].value_counts().head(1)
print("The most common rating category is:",mcommon_rating)
```

```python
#Calculate the distribution of content by country for the top 5 countries.

top5_country=df["country"].value_counts().head(5)
print("The top 5 countries by content distribution:",top5_country)
```

```python
#to analyze content types (Movie vs TV Show) and their average release year distribution.

content_avg=df.groupby("type")["release_year"].mean()
print("The content types and their average release year distribution is:",content_avg)
```

#4. Trend and Relationship Exploration

```python
#to identify how many titles were added to Netflix in each year.

titles_peryear=df.groupby("Year")["show_id"].count()
print("The titles that were added to netflix in each year:",titles_peryear)
```

Determine if there is any noticeable trend in the number of titles added per year.

To understand how Netflix's content library has evolved over time, the number of titles added each year from the date_added column and counting the number of entries per year. The analysis reveals a clear upward trend in content additions over time. Prior to 2015, the number of titles added each year was relatively low, indicating that Netflix had a smaller content library and a more limited global presence during its early years. Starting around 2015, there is a noticeable and consistent increase in the number of titles added annually. This period marks a phase of rapid growth. The increase continues sharply through 2016,2017 and 2018. The number of titles added appears to peak around 2019-2020, indicating the height of Netflix's expansion phase.

```python
#Analyze the relationship between release year and rating type visually.

#using scatter plot
plt.scatter(df['release_year'],df['rating'])
plt.title('release year vs rating')
plt.xlabel('release year')
plt.ylabel('rating')
plt.show()


```

#5. Visualizations

```python
#count plot for type (Movies vs TV Shows).

sns.countplot(x="type",data=df)
plt.title("Movies vs TV shows")
plt.show()
```

```python
#bar chart showing the top 5 countries by content count.

top5_country.plot(kind="bar")
plt.title("Top 5 countries by content")
plt.show()
```

```python
#Create a line chart showing yearly title additions.

plt.figure()
titles_peryear.plot()
plt.title("Titles added per year")
plt.xlabel("Year")
plt.ylabel("Title")
plt.show()
```

```python
#scatter plot of release year vs duration.
#to convert duration column to numeric

df["duration_num"]=df["duration"].str.split().str[0].astype(int)

#scatter plot

plt.scatter(df["release_year"],df["duration_num"])
plt.xlabel("Release year")
plt.ylabel("Duration")
plt.title("Release year vs Duration")
plt.show()
```

```python
#Generate a heatmap of correlations between numeric features.

corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Summarize key findings about Netflix content trends.


The exploratory data analysis of the Netflix dataset reveals several important trends in the platform's content strategy and evolution over time.

2)Dominance of movies

OVERALL INSIGHT

Netflix has evolved from a smaller content provider into a global streaming platform with rapid content expansion. The analysis indicates that Netflix's strategy has shifter from content quantity expansion to a more balanced approach emphasising quality,diversity and global reach.
